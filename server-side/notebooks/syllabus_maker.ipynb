## List to string conversion

```python
def list_to_string(list):  
    string = ''
    for item in list:
        string += item + ', '
    return string[:-2]
# end of list_to_string 

skills_str = list_to_string(skills)
```

## Prompt Generation

```python
def generate_prompt(section, purpose, comp_name, job_title, job_desc, type, topic_title, skills, target_skills, industry, current_job, target_industry) -> str:
    
    if section == 'Syllabus':
        if purpose == 'Upskill':
            if type == 'topic':
                if not industry:
                    prompt = """Act as a professional syllabus maker. create a numerical bulleted list of titles for a syllabus pathway to attain the skills and knowledge needed to attain the {} skill.""".format(target_skills)
                
                else:
                    prompt = """Act as a professional syllabus maker. create a numerical bulleted list of titles for a syllabus pathway to attain the skills and knowledge needed, in the {} field, to attain the {} skill.""".format(industry, target_skills)
                
            elif type == 'subtopic':
                if not industry:
                    prompt = """Act as a professional syllabus maker. create a list of sub titles for the title "{}" for a syllabus pathway to attain the skills and knowledge needed to attain the {} skill. The subtopic should be indented and bulleted by a lowercase letters, while the main topic should retain its bulleting format which is numerical.""".format(topic_title, target_skills)
                
                else:
                    prompt = """Act as a professional syllabus maker. create a list of sub titles for the title "{}" for a syllabus pathway to attain the skills and knowledge needed, in the {} field, to attain the {} skill. The subtopic should be indented and bulleted by a lowercase letters[a-z], while the main topic should retain its bulleting format which is numerical.""".format(topic_title, industry, target_skills)
        # end of upskill 
                    
        elif purpose == 'Job':
            if type == 'topic':
                if not skills:
                    prompt = """{}
                    Act as a professional syllabus maker. create a numerical bulleted list of titles for a syllabus pathway to attain the skills and knowledge mentioned in the job description for {} in {}.""".format(job_desc, job_title, comp_name)
                else:
                    prompt = """{}
                    Act as a professional syllabus maker. create a  numerical bulleted list of titles for a syllabus pathway to attain the skills and knowledge mentioned in the job description for {} in {}. Exclude from the syllabus the following skills: {}""".format(job_desc, job_title, comp_name, skills)
                
            elif type == 'subtopic':
                prompt = """{}
                act as a professional syllabus maker. create a list of sub titles for the title "{}" for a syllabus pathway to attain the skills and knowledge mentioned in the job description for {} in {}. The subtopic should be indented and bulleted by a lowercase letters[a-z], while the main topic should retain its bulleting format which is numerical.""".format(job_desc, topic_title, job_title, comp_name)
        # end of job
        
        elif purpose == 'Career':
            if type == 'topic':
                if not industry:
                    prompt = """Act as a professional syllabus maker. create a numerical bulleted list of titles for a syllabus pathway to attain the skills and knowledge needed to attain the skill to switch from {} to {} job.""".format(current_job, job_title)
                
                else:
                    prompt = """Act as a professional syllabus maker. create a numerical bulleted list of titles for a syllabus pathway to attain the skills and knowledge needed to attain the skill to switch from {} ({} field) to {} ({} field) job.""".format(current_job, industry, job_title, target_industry)
                
            elif type == 'subtopic':
                if not industry:
                    prompt = """Act as a professional syllabus maker. create a list of sub titles for the title "{}" for a syllabus pathway to attain the skills and knowledge needed to attain the skill to switch from {} to {} job. The subtopic should be indented and bulleted by a lowercase letters, while the main topic should retain its bulleting format which is numerical.""".format(topic_title, current_job, job_title)
                
                else:
                    prompt = """Act as a professional syllabus maker. create a list of sub titles for the title "{}" for a syllabus pathway to attain the skills and knowledge neededto attain the skill to switch from {} ({} field) to {} ({} field) job.. The subtopic should be indented and bulleted by a lowercase letters[a-z], while the main topic should retain its bulleting format which is numerical.""".format(topic_title, current_job, industry, job_title, target_industry)
        # end of Career
        
        # For adding the skills, company name, and job description to the prompt if they exist 
        if skills:
            add_prompt = """ The syllabus should exclude the following skills: {}.""".format(skills_str)
            prompt = prompt + add_prompt
            
        if comp_name:
            add_prompt = """ Then, the syllabus should also for landing a job in the {} company.""".format(comp_name)
            prompt = prompt + add_prompt
            
        if job_desc:
            add_prompt = """{}
            """.format(job_desc)
            
            prompt = add_prompt + prompt
            add_prompt = """ Then, the syllabus should also be focusing on the skills and knowledge needed in the job description mentioned above."""
            prompt = prompt + add_prompt
    # end of Syllabus section
            
    return prompt
# end of generate_prompt
```

## Response Generation

```python
def generate_response(prompt) -> str:
    response = openai.Completion.create(
                model="text-davinci-003",
                prompt=prompt,
                temperature=0.1,
                max_tokens=1000,
            )
    
    return response
# end of generate_response
```

```python
prompt = generate_prompt('Syllabus', 'Upskill', target_company, desired_job, job_desc, 'topic', '', skills_str, target_skills, industry, current_job, target_industry)
topics_list = generate_response(prompt)
```

## List cleaning

```python
def clean_lst(lst) -> list:
    lst = lst.choices[0].text
    lst = lst.split('\n')
    lst = [x for x in lst if x]
    
    return lst
# end of clean_lst
    
topics_list = clean_lst(topics_list)
```

## Keeping only the bulleted responses

```python
def keep_bullets_and_numbers(lst):
    pattern = r'^[\s*]*[\d|a-zA-Z]\.|[\s*]*[-–]\s+'
    
    return [s for s in lst if re.match(pattern, s)]
# end of keep_bullets_and_numbers

topics_list = keep_bullets_and_numbers(topics_list)
topics_list
```

## Syllabus Generation

```python
def generate_syllabus(topics_list):
    syllabus = []

    for topic in topics_list:
        prompt = generate_prompt('Syllabus', 'Upskill', target_company, desired_job, job_desc, 'subtopic', topic, skills_str, target_skills, industry, current_job, target_industry)
        # prompt = generate_prompt(target_company, desired_job, job_desc, 'subtopic', topic, '')
        subtopic = generate_response(prompt)
        subtopic = clean_lst(subtopic)
        syllabus.append(subtopic)
    # end of for loop
    
    return syllabus
# end of generate_syllabus
    
syllabus = generate_syllabus(topics_list)
syllabus
```
